```python
class PLL():
    def __init__(self,initial_phase=0.1,lr=1):
        self.phase = initial_phase
        self.VCO = np.exp(1j*(self.phase))
        self.lr = lr
    def LF(self, phase_diff):
        return self.lr*phase_diff
    def forward(self,x):
        phase_diff = np.angle(x*np.conj(self.VCO))
        return self.LF(phase_diff)
        
    
```

```python
pll0 = PLL()
pll0.forward(x_in[2])
```

```python
pll0 = PLL(initial_phase=0.7*PI, lr=3)
losses = []
for i in x_in:
    loss = pll0.forward(i)
    pll0.phase = pll0.phase + loss
    losses.append(pll0.phase)
    # print(loss)

plt.plot(losses)
```
